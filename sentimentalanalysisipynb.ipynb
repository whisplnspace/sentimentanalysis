```python


# Step 2: Import the necessary libraries
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt

# Step 3: Load the image
# If you have an image uploaded, you can use cv2.imread.
# Alternatively, you can use an image URL.

image_path = '/content/angry2.jpg'  # Replace with the path to your image

# Load the image using OpenCV
image = cv2.imread(image_path)

# Convert the image from BGR (OpenCV format) to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb)
plt.axis('off')
plt.show()

# Step 4: Perform emotion analysis using DeepFace
results = DeepFace.analyze(image_rgb, actions=['emotion'], enforce_detection=False)

# Step 5: Display the results
# Iterate through the results list if multiple faces were detected
for result in results:
    print("Detected Emotion: ", result['dominant_emotion'])
    print("All Emotion Scores: ", result['emotion'])
```
